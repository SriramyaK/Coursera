```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np  # useful for many scientific computing in Python
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
r = requests.get('https://en.wikipedia.org/wiki/Visakhapatnam')
soup = BeautifulSoup(r.text,'html.parser')
results = soup.find_all('p')
records = []
for result in results:
    records.append(result.text.strip('\n'))
#print(str(records))
str1 = ''.join(records)
str1
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
```

```python
Vizag_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
Vizag_wc.generate(str1)
```

```python
plt.imshow(Vizag_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(Vizag_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('Buddhist') # add the words said to stopwords

# re-generate the word cloud
Vizag_wc.generate(str1)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(Vizag_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
Vizag_mask = np.array(Image.open('vishakapatnam_mandal_map.png'))
    
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(Vizag_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
Vizag_wc = WordCloud(background_color='white', max_words=2000, mask=Vizag_mask, stopwords=stopwords)

# generate the word cloud
Vizag_wc.generate(str1)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(Vizag_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
